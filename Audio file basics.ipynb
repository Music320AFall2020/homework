### Audio file processing
[Librosa](https://librosa.org/) is a audio library that is used for audio signal processing. 

```python
import numpy, scipy, matplotlib.pyplot as plt, IPython.display as ipd
import librosa, librosa.display
```

```python
x, sr = librosa.load('audio/85bpm.wav')
ipd.Audio(x, rate=sr)
print(len(x), sr)
```

```python
# Display waveform
librosa.display.waveplot(x, sr, alpha=0.8)
```

```python
# Determine frequency using fft 
X = scipy.fft.fft(x[:4096])
X_mag = numpy.absolute(X)        # spectral magnitude
f = numpy.linspace(0, sr, 4096)  # frequency variable
plt.figure(figsize=(14, 5))
plt.plot(f, X_mag) # magnitude spectrum
plt.xlabel('Frequency (Hz)')
```

```python
t = numpy.linspace(0,2,44100)
x = numpy.sin(2*numpy.pi*1000*t)

# Plot only the first 100 samples so you can see waveform
plt.plot(t[:100], x[:100])
ipd.Audio(x, rate=22050)
```
